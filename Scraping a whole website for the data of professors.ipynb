```python
import bs4

from bs4 import BeautifulSoup

import requests
```

```python
html_text=requests.get('https://www.swansea.ac.uk/staff/engineering/#associate-professors=is-expanded&lecturers-and-tutors=is-expanded&professors=is-expanded&readers=is-expanded&senior-lecturers=is-expanded').text
```

```python
soup=bs4.BeautifulSoup(html_text,'html')
```

```python
page_1_academic_staff_profile_link=soup.find('div',class_='title-and-body-text title-and-body-text-12')
```

```python
my_dict={}
```

```python
for i in [1,2,3,4,5]:
    site=page_1_academic_staff_profile_link.find_all('p')[i*2].a['href']
    
    
    prof_site=requests.get(site).text
    soup=bs4.BeautifulSoup(prof_site,'html')
    
    
    name=soup.find('h1',class_="staff-profile-overview-honorific-prefix-and-full-name").text.replace(' ','_')
    print(name)
    
    my_dict[name]={}
    print(1)
    my_dict[name]['about']=[]
    for i in range(0,len(soup.find('div',class_="title-and-body-text title-and-body-text-12").find_all('p'))):
        about=soup.find('div',class_="title-and-body-text title-and-body-text-12").find_all('p')[i].text
        my_dict[name]['about'].append(about)
        
    
    
    
    
    print(2)
    #####  Telephone,email,research_link
    
    my_dict[name]['Phone_no']=[]
    my_dict[name]['email']=[]
    my_dict[name]['research_link']=[]
    
    for i in range(0,len(soup.find('div',class_="col-12 col-lg-6").find_all('div',class_="col"))):
        try:
            if i==0:
                tele=soup.find('div',class_="col-12 col-lg-6").find('div',class_="col").div.text
                my_dict[name]['Phone_no'].append(tele)
            
            
            else:
                email=soup.find('div',class_="col-12 col-lg-6").find_all('div',class_="col")[i].div.text.replace('\n','').replace('  ','')
                if email=='JavaScript is required to view this email address.':
                    continue
                my_dict[name]['email'].append(email)
                
                
            
        except AttributeError:
            research_link=soup.find('div',class_="col-12 col-lg-6").find_all('div',class_="col")[i].a['href']
            my_dict[name]['research_link'].append(research_link)
            
            
    
    print(3)
    #########available_for,position
    my_dict[name]['available_for']=[]
    my_dict[name]['position']=[]
            
    for i in range(0,len(soup.find_all('div',class_="col-12 col-lg-6")[1].find_all('div',class_='col'))):
        
        
        if len(soup.find_all('div',class_="col-12 col-lg-6")[1].find_all('div',class_='col'))==1 or i==1:
               
               my_dict[name]['available_for']=soup.find_all('div',class_="col-12 col-lg-6")[1].find_all('div',class_='col')[i].text.replace('\n','').replace('  ','')
        
               
        
        else:
               my_dict[name]['position']=soup.find_all('div',class_="col-12 col-lg-6")[1].find_all('div',class_='col')[0].text.replace('\n',',').replace('  ','')[1:-1]

               
            
             
    print(4)            
    my_dict[name]['area_of_expertise']=[]
    try:
        for i in range(0,len(soup.find('div',class_="staff-profile-areas-of-expertise").find_all('li'))):
            my_dict[name]['area_of_expertise'].append(soup.find('div',class_="staff-profile-areas-of-expertise").find_all('li')[i].text)
            
            
            
            
    except AttributeError:
        pass
        
        
        
        
    print(5)    
    my_dict[name]['teaching_interests']=[]
    my_dict[name]['award_highlights']=[]
    my_dict[name]['collaborations']=[]
    my_dict[name]['research']=[]    
    j=0
    try:
        
        for i in range(0,len(soup.find('div',class_="tabs su-tabs").find_all('button',class_="su-nav-item nav-item"))):
            
            try:
                len(soup.find('div',class_="tabs su-tabs").find_all('div',class_="tab-content-panel")[i].img)
                
            except TypeError:
                
                
                header=soup.find('div',class_="tabs su-tabs").find_all('button',class_="su-nav-item nav-item")[i].text.replace('\n','').replace('  ','').replace(' ','_').lower()
                my_dict[name][header].append(soup.find('div',class_="tabs su-tabs").find_all('div',class_="col-md-12")[j].text.replace('\n',',')[1:-1])
                j=j+1
              
            
         
        
        
        
    except AttributeError:
        pass
            
        
    my_dict[name]['supervising']=[]    
    ############ Supervising    
    print(6)
    try:
        for i in range(0,len(soup.find('div',class_="article-item expander-list").find_all('h4',class_="staff-profile-supervision-expander-thesis-title")[0])):
            title=soup.find('div',class_="article-item expander-list").find_all('h4',class_="staff-profile-supervision-expander-thesis-title")[i].text.replace('\n','').replace('  ','')
            course=soup.find('div',class_="article-item expander-list").find_all('div',class_="staff-profile-supervision-expander-thesis-type")[i].text.replace('\n','').replace('  ','')
            other_supervisor=soup.find('div',class_="article-item expander-list").find_all('div',class_="staff-profile-supervision-expander-thesis-other-supervisor")[i].text.replace('\n','   ').replace('  ','')
             
        
            my_list=[title,course,other_supervisor]
            
        
            my_dict[name]['supervising'].append(my_list)
            
        
    except IndexError:
        
        pass
    
    
    
    
    print(7)
    ##############course_taught
    my_dict[name]['courses_taught']=[]
    try:
        for i in range(0,len(soup.find('div',class_="article-item expander-list").find_all('h4',class_="staff-profile-taught-modules-expander-module-title"))):
            
            my_dict[name]['courses_taught'].append(soup.find('div',class_="article-item expander-list").find_all('h4',class_="staff-profile-taught-modules-expander-module-title")[i].text.replace('\n',',').replace('  ','')[1:-1])
    
    
     
    except IndexError:
        pass
    
    
    
    
    print(8)
    ##################key_grant_&_projects
    my_dict[name]['key_grant_&_projects']=[]
    try:
        for i in range(0,len(soup.find_all('li',class_="staff-profile-key-grants-and-projects-expander-list-item"))):
                       title=soup.find_all('li',class_="staff-profile-key-grants-and-projects-expander-list-item")[i].h4.text.replace('\n',',').replace('  ','')[1:-1]
                       details=soup.find_all('li',class_="staff-profile-key-grants-and-projects-expander-list-item")[i].p.text.replace('\n',',').replace('  ','')[1:-1]
                       
                       final=title+':'+details
                       
                       my_dict[name]['key_grant_&_projects'].append(final)
                       
                       
                       
                       
                       
    except IndexError:
                       pass
    
    
    
    
    print(9)
    ##############affliated_positions
    my_dict[name]['affliated_positions']=[]
    try:
        for i in range(0,len(soup.find_all('li',class_="staff-profile-affiliated-positions-expander-list-item"))):
            position=soup.find_all('li',class_="staff-profile-affiliated-positions-expander-list-item")[0].h4.text.replace('\n',',').replace('  ','')[1:-1]
            time=soup.find_all('li',class_="staff-profile-affiliated-positions-expander-list-item")[0].p.text.replace('\n',',').replace('  ','')[1:]
            final=position+','+time
            print(final)
            my_dict[name]['affliated_positions'].append(final)
            
            
            
            
    except IndexError:
        pass
    
    
    print(10)
    ##########################3publication
    my_dict[name]['publication']=[]
    try:
        for i in range(0,len(soup.find_all('ol',class_="staff-profile-publications-expander-publications"))):
            for j in range(0,len(soup.find_all('ol',class_="staff-profile-publications-expander-publications")[i].find_all('li'))):
                publication=soup.find_all('ol',class_="staff-profile-publications-expander-publications")[i].find_all('li')[j].div.text
                my_dict[name]['publication']=publication
                print(publication)
                
                
                
    except IndexError:
        pass
    
               
               
            
            

            
        
        
        
        
    
            
    
    
    
    
    
    
```
